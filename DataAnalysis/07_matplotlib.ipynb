## 07. matplotlib
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리!

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.__version__

plt.rcParams['font.family'] = 'Malgun Gothic'   # 맑은 고딕이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False      # 마이너스(-) 부호 깨짐 방지
```

## 그래프 그리기

```python
x = [1,2,3,4,5]
y = [10, 13, 17, 19, 24]

# 선 그래프
plt.plot(x,y)

# 이미지를 파일로 저장(그리기 전에 먼저 해야함)
plt.savefig("my_graph.png")

# 콘솔에 이미지 그리기
plt.show()  
```

### 그래프 구성요소(1)
- figure : 여러 그래프가 그려지는 최상위 컨테이너
- axes : 실제 그래프가 그려지는 공간
- axis : 그래프의 축

```python
# subplots()
# - 여러개의 그래프 그릴 때 사용
# - figure, axes 동시에 반환
# fig, axs = plt.subplots(행, 열)
fig, axs = plt.subplots()   # 하나의 figure와 하나의 axes 만듦
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1,2)

axs[0].plot([1,2,3], [1,2,3])
axs[1].plot([1,2,3], [1,4,9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2,2)
a = [1,2,3]
b = [1,4,9]

fig.suptitle("여러 함수 그래프")

axs[0, 0].plot(a,a)
axs[0, 0].set_title("1번 그래프")

axs[0, 1].plot(a,b)
axs[0, 1].set_title("2번 그래프")

axs[1, 0].plot(a,[3,2,1])
axs[1, 0].set_title("3번 그래프")

axs[1, 1].plot(a,[9,4,1])
axs[1, 1].set_title("4번 그래프")

plt.tight_layout()  # 그래프 간격 자동 조정

plt.show()
```

### 그래프의 구성요소(2)
- 제목, 레이블, 범례, 그리드

```python
# title
plt.plot([1,2,3,4,5], [1,4,9,16,25], label='제곱 그래프', linestyle='--', color="c", marker="^")
plt.title("제곱 함수 그래프", color="b", fontsize=20)

# xlabel, ylabel
plt.xlabel("x값")
plt.ylabel("y값(x의 제곱)")

plt.show()  
```

```python
# title
plt.plot([1,2,3,4,5], [1,4,9,16,25], label='제곱 그래프', linestyle='--', color="c", marker="^")
plt.title("제곱 함수 그래프", color="b", fontsize=20)

# xlabel, ylabel
plt.xlabel("x값")
plt.ylabel("y값(x의 제곱)")

# 범례 (legend)
plt.legend()

# 격자 (grid)
plt.grid(linestyle="--")

plt.show()  
```

```python
# 하나의 axes에 여러개의 그래프 그리기
x = np.arange(0,10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker='o', label="y = 2x + 1", color="c")
plt.plot(x, y2, marker="*", label="y = 3x - 5", color="m")

plt.title("두개의 선 그래프 비교")
plt.xlabel("x값")
plt.ylabel("y값")
plt.legend()
plt.grid(linestyle=":", alpha=0.5)

plt.show()
```

```python
# <실습1>
# 아래 데이터를 이용해서 예시와 같이 선 그래프를 그려보세요
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, marker='o', label="2019매출", color="b")
plt.plot(months, sales_2020, marker="^", label="2020매출", color="r")

plt.title("Monthly Sales Comparison(2019-2020)")
plt.xlabel("Monthly", labelpad=15)
plt.ylabel("Sales", labelpad=15)
plt.legend()
plt.grid(linestyle=":", alpha=0.5)
plt.legend(loc="lower right", title="판매 연도")
plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기", "2분기", "3분기", "4분기"]
men_means = [20,34,30,35]
women_means = [25,32,34,40]

x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.4                     # 막대 너비

ax.bar(x - width/2, men_means, width, label = "남성")
ax.bar(x + width/2, women_means, width, label = "여성")

ax.set_xticks(x, labels)        # 숫자 대신 문자열을 표시
plt.legend()
plt.show()
```

```python
# <실습2>
# 아래 데이터를 바탕으로 예시와 같이 막대 그래프를 그려보세요
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

plt.bar(categories, data)
plt.title("Bar Chart")
plt.grid(linestyle=":", alpha=1)

plt.xlabel("Categories",labelpad=15)
plt.ylabel("Values", labelpad=15)

plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.ylim(0,50)                  # y축을 0 ~ 50 범위로 고정

plt.show()
```

### 여기 밑으로는 수업과는 예외

```python
plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')

# make data
x = np.linspace(-4, 4, 6)
y = np.linspace(-4, 4, 6)
X, Y = np.meshgrid(x, y)
U = X + Y
V = Y - X

# plot
fig, ax = plt.subplots()

ax.quiver(X, Y, U, V, color="C0", angles='xy',
          scale_units='xy', scale=5, width=.015)

ax.set(xlim=(-5, 5), ylim=(-5, 5))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# Make data for a double helix
n = 50
theta = np.linspace(0, 2*np.pi, n)
x1 = np.cos(theta)
y1 = np.sin(theta)
z1 = np.linspace(0, 1, n)
x2 = np.cos(theta + np.pi)
y2 = np.sin(theta + np.pi)
z2 = z1

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.fill_between(x1, y1, z1, x2, y2, z2, alpha=0.3)
ax.plot(x1, y1, z1, linewidth=2, color='C0')
ax.plot(x2, y2, z2, linewidth=2, color='C0')

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
t = np.linspace(0, 10*np.pi, 500)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z)
ax.set_title("3D Helix (cos(t), sin(t), t)")
plt.show()
```

```python
plt.style.use('_mpl-gallery')

n = 200
t = np.linspace(0, 6*np.pi, n)

x1 = np.cos(t)
y1 = np.sin(t)
z1 = np.sin(t) * 0.3 + t / (6*np.pi)

x2 = np.cos(t + np.pi)
y2 = np.sin(t + np.pi)
z2 = np.sin(t + np.pi) * 0.3 + t / (6*np.pi)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.plot(x1, y1, z1, color='C0', linewidth=2)
ax.plot(x2, y2, z2, color='C0', linewidth=2)
ax.plot_surface(
    np.vstack([x1, x2]),
    np.vstack([y1, y2]),
    np.vstack([z1, z2]),
    alpha=0.3,
    color='C0'
)

ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
plt.show()
```
