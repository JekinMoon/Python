```python
# 스타일 설정

sns.set_style("ticks")
sns.set_context("notebook")         # 문맥에 따라 스타일 크기 조정
sns.set_palette("pastel")

plt.rcParams['font.family'] = 'Malgun Gothic'   # 맑은 고딕이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False      # 마이너스(-) 부호 깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data=tips, x='day', hue="sex", palette='pastel')
plt.title("요일별 방문자수")
plt.grid(linestyle=":")
plt.show()
```

```python
# barplot

palette=sns.color_palette("colorblind")
sns.barplot(data=tips, x="day", y="tip", palette=palette)
plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# <gpt 꺼>

# seaborn 내장 데이터셋 불러오기
tips = sns.load_dataset("tips")

# 색상 팔레트
palette = sns.color_palette("colorblind")

plt.figure(figsize=(8,5))

# 1️⃣ 막대 그래프: 평균 + 95% 신뢰구간
sns.barplot(data=tips, x="day", y="tip", palette=palette, ci=95)

# 2️⃣ 실제 데이터 점 표시
sns.stripplot(data=tips, x="day", y="tip", color='black', size=5, jitter=True, alpha=0.7)

plt.title("요일별 평균 팁 금액과 신뢰구간 + 데이터 점")
plt.show()
```

```python
# displot()
# 히스토그램 + 커널 밀도(kde)
# kde = True 로 하면 kde 그래프 그려준다
sns.displot(data=tips, x="total_bill", bins=20)
sns.displot(data=tips, x="total_bill", kde=True, bins=20)
plt.title("Total Bill")
plt.show()
```

```python
# scatterplot()
# 산점도

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
# countplot
palette = sns.color_palette("pastel")
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x='class', hue="sex", palette='pastel')
plt.title("클래스별 남/녀 승객수")
plt.grid(linestyle=":")
plt.show()

# barplot
palette=sns.color_palette("colorblind")
sns.barplot(data=titanic, x="class", y="age", hue="sex", palette='pastel')
plt.title("클래스별 남/녀 나이대")
plt.grid(linestyle=":")
plt.show()

# displot()
plt.figure(figsize=(8,5))
sns.displot(data=titanic, x="fare", bins=20, kde=True)
plt.title("요금")
plt.grid(linestyle=":")
plt.show()

# scatterplot()
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex")
plt.title("객실 등급별 승객 나이(성별 구분)")
plt.grid(linestyle=":")
plt.show()
```

```python
# heatmap()
# 상관계수행렬의 그래프
penguins = sns.load_dataset("penguins")
print(penguins.head(10))
print(penguins["species"].unique())

# 상관계수행렬
# - 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# - heatmap에서 색으로 표시되는 숫자가 상관계수가 된다

# 상관계수 1 : 양의 상관관계(하나가 증가하면 다른 것도 같은 비율로 증가)
# ex) 키와 몸무게

# 상관계수 -1 : 음의 상관관계 (하나가 감소하면 다른 변수도 감소)
# ex) 공부시간과 실수 개수

corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="inferno")
plt.show()
```

```python
# jointplot()
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

# heatmap()
species = penguins.groupby("species").mean(numeric_only=True)
plt.figure(figsize=(8,4))
sns.heatmap(species, annot=True, fmt=".2f", cmap="magma")
plt.title("펭귄 종류별 평균 신체")
plt.show()

# jointplot()
sns.jointplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="species", kind="scatter")
plt.suptitle("펭귄 종별 부리길이와 신체")
plt.show()
```
